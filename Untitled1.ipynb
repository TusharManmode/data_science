```python
import pandas as pd
from sklearn import preprocessing
df = pd.read_csv("D:/tusharpythonnn/data_science/WA_Fn-UseC_-Sales-Win-Loss.csv")
le = preprocessing.LabelEncoder()
df.head()
```

```python
feature_cols = [col for col in df.columns if col not in ['Opportunity Number','Opportunity Result']]
target = df['Opportunity Result']
data = df[feature_cols]
data.head()
```

```python
df['Supplies Subgroup']=le.fit_transform(df['Supplies Subgroup'])
df['Region']=le.fit_transform(df['Region'])
df['Route To Market']=le.fit_transform(df['Route To Market'])
df['Opportunity Result']=le.fit_transform(df['Opportunity Result'])
df['Competitor Type']=le.fit_transform(df['Competitor Type'])
df['Supplies Group']=le.fit_transform(df['Supplies Group'])
```

```python
feature_cols=[col for col in df.columns if col not in ['Opportunity Number','Opportunity Result']]
target=df['Opportunity Result']
data=df[feature_cols]
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.30, random_state = 10)
print(x_train)
print(y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
gnb = GaussianNB()
pred = gnb.fit(x_train,y_train).predict(x_test)
print(pred)
accuracy_score(y_test,pred,normalize=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
neigh=KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
pred=neigh.predict(x_test)
accuracy_score(y_test,pred,normalize=True)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
svc_model=LinearSVC(random_state=1,max_iter=100)
pred=svc_model.fit(x_train, y_train).predict(x_test)
print(pred)
accuracy_score(y_test, pred, normalize=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

#data =np.array(df['Supplies Subgroup']),reshape(-1,1)
x_train,x_test,y_train,y_test=train_test_split(data,target,test_size=0.30,random_state=10)
reg=LinearRegression()
reg.fit(x_train, y_train)
pred=reg.predict(x_test)
print("pred===",pred)
#accuracy_score(y_test,pred,normalize=True)
#print(reg.score(y_test,pred))
print("reg.coef_==",reg.coef_)
print("reg.intercept_==",reg.intercept_)
print("len(x_test)==",len(x_test))
print("y_test==",len(y_test))
plt.scatter(x_test, y_test, color='gray')
plt.plot(y_test,pred,color='red',linewidth=2)
plt.show()
```
